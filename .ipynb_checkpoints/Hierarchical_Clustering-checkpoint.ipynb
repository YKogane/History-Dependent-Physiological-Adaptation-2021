# Program used for hierarchical clustering

This program is used in Figure S10A and B.

### import packages used in this script

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import random
import os
from fastdtw import fastdtw
from scipy.cluster.hierarchy import linkage, dendrogram
import scipy.spatial.distance as distance
from common_function import singlecell as sc
```

### Matplotlib setting

```python
plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = [r'\usepackage{sansmath}', r'\sansmath'] 
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

plt.rcParams["font.size"] = 16

plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
```

### set valiables

```python
path_list = ['./single-cell_data/deletion_in_Cp/180726']
branch_path = path_list[0]
BL_time_Cp_time_list = pd.read_csv('./BL_time_Cp_time_data.csv')
BL_time = BL_time_Cp_time_list.loc[BL_time_Cp_time_list['Date'] == int(branch_path[-6:]), 'BL_time'].reset_index(drop = True)[0] ## slice number just after blue light exposure
Cp_time = BL_time_Cp_time_list.loc[BL_time_Cp_time_list['Date'] == int(branch_path[-6:]), 'Cp_time'].reset_index(drop = True)[0] ## slice number just after Cp exposure
if BL_time == 0:
    BL_time = Cp_time
window = 1/12 ## averaged time window (1/12 is non-averaged window)
```

## Figure S10A

### load single-cell result data in deletion expeiment.

```python
image_path = branch_path
Number = os.listdir(image_path + '/integrated')
Number = [i for i in Number if 'Results' in i]
for i in range(len(Number)):
    Number[i] = Number[i].lstrip('Results')
    Number[i] = Number[i].rstrip('.csv')
Result_all = sc.Read_results_data(Number, dr = image_path)      
```

### calculate distant matrix

Distance between two single-cell mCherry-CAT transitions are calculated using dynamic time warping divided by path length.

```python
dist_ave_matrix = []

for i in np.arange(0,len(Number),1):
    Result1 = Result_all.loc[Result_all['file'] == int(Number[i]),'TxRed'].dropna()
    for j in np.arange(i+1,len(Number),1):
        Result2 = Result_all.loc[Result_all['file'] == int(Number[j]),'TxRed'].dropna()
        distance, path = fastdtw(Result1,Result2)
        dist_ave_matrix += [distance/len(path)]
```

### calculate the linkage and plot dendrogram

This hierarchical clustering is aggromerative.  
Distance between groups or a group and a single-cell transition is used in averaged distance of all component including the groups.

```python
###clustering
result = linkage(dist_ave_matrix, method = 'average')
 
###show figure
plt.figure(figsize = (20,12))
plt.ylabel('Dynamic time warping', fontsize = 18)
plt.xlabel('Result No.', fontsize = 18)
plt.yticks([0,10,20,30,40,50,60,70],[0,'',20,'',40,'',60,''])
dendrogram(result,labels = Number, leaf_font_size=14)
plt.show()
#save_path = './FigureS10A'
#plt.savefig(save_path, transparent=True, bbox_inches = 'tight')
#plt.cla()
```

## Figure S10B

### load single-cell result data in deletion expeiment and calculate elongation rates.

```python
value = 'elongation rate'
win = 1.0
Result_all = pd.DataFrame(data = {'Time':[],value:[],'file':[]})
for i in Number:
    Result = pd.read_csv('./single-cell_data/deletion_in_Cp/180726/integrated/Results'+str(i).zfill(4)+'.csv', encoding = 'shift_jis')
    Result['Time'] = (Result['Slice']-BL_time)*10/60
    Result['Time2'] = Result['Time'].shift(1)
    Result['Area2'] = Result['Area'].shift(1)
    Result[value] = np.log(Result['Area2']/Result['Area'])/(Result['Time2']-Result['Time'])
    Result.loc[Result[value] <= np.log(0.75)/(Result['Time2']-Result['Time']), value]=np.nan
    time = []
    values = []
    files = []
    for t in np.arange(-1,125,2*win):
        time += [t]
        files += [int(i)]
        values += [Result[(Result['Time']>=t-win)&(Result['Time']<t+win)][value].mean()]
    Result_all = pd.concat([Result_all, pd.DataFrame(data = {'Time':time,value:values,'file':files})])
```

### calculate distant matrix

Distance between two single-cell transitions of elongation rates is calculated using dynamic time warping divided by path length.

```python
dist_ave_matrix = []

for i in np.arange(0,len(Number),1):
    Result1 = Result_all.loc[Result_all['file'] == int(Number[i]),value].dropna()
    for j in np.arange(i+1,len(Number),1):
        Result2 = Result_all.loc[Result_all['file'] == int(Number[j]),value].dropna()
        distance, path = fastdtw(Result1,Result2)
        dist_ave_matrix += [distance/len(path)]
```

### calculate the linkage and plot dendrogram

This hierarchical clustering is aggromerative.  
Distance between groups or a group and a single-cell transition are used in averaged distance of all component including the groups.

```python
###clustering
result = linkage(dist_ave_matrix, method = 'average')
 
###show figure
plt.figure(figsize = (20,12))
plt.ylabel('Dynamic time warping', fontsize = 18)
plt.xlabel('Result No.', fontsize = 18)
plt.yticks([0,0.025,0.05,0.075,0.10,0.125,0.150,0.175,0.200],[0,'',0.05,'',0.10,'',0.15,'',0.20])
dendrogram(result,labels = Number, leaf_font_size=14)
plt.show()
#save_path = './FigureS10B'
#plt.savefig(save_path, transparent=True, bbox_inches = 'tight')
#plt.cla()
```
