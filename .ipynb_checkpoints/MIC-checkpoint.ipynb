# Programs for figures relate to MIC test.

This program is used in Figure S3 and S9.

### import packages

```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

### Matplotlib setting

```python
plt.rcParams['font.size'] = 16

plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = [r'\usepackage{sansmath}', r'\sansmath'] 
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

plt.rcParams["font.size"] = 16

plt.rcParams['xtick.direction'] = 'out'#x軸の目盛線が内向き('in')か外向き('out')か双方向か('inout')
plt.rcParams['ytick.direction'] = 'out'#y軸の目盛線が内向き('in')か外向き('out')か双方向か('inout')
```

## Figure S3

MIC tests of YK0083, YK0085 and their parental strain, F3 (W3110 _∆fliC_ _∆fimA_ _∆flu_).

### extract the strain name

```python
main_path = '.'
branch_path = '/MIC_data/FigureS3'
folder_path = main_path + branch_path

Folder_list=os.listdir(folder_path)
if '.DS_Store' in Folder_list:
    Folder_list.remove('.DS_Store')
Folder_list[1], Folder_list[2] = Folder_list[2], Folder_list[1] ### change the order to ['YK0083', 'YK0085', 'F3']
```

### import the MIC results and plot figures

```python
min_func = lambda x: max(x, 0.01) ### constructed function for map function
fig, ax = plt.subplots(ncols = 3, nrows = 1,figsize = (12,3), sharey = True)
for i, strain in enumerate(Folder_list):
    data_path = folder_path + '/' + strain
    Day_list = os.listdir(data_path)
    strain_total = pd.DataFrame({'Cp_con':[], strain:[]})
    for day in Day_list:
        day_data = pd.read_csv(data_path + '/' + day)
        strain_total = pd.concat([strain_total, day_data])
    strain_average = strain_total.groupby('Cp_con', as_index = False).mean()
    strain_total[strain] = strain_total[strain].map(min_func)
    ax[i].set_xlim(0.1,1000)
    ax[i].set_ylim(0.01,2)
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')
    ax[i].plot(strain_average['Cp_con'], strain_average[strain], color = 'black', label = strain, lw =2) ### draw black lines of average OD data.
    ax[i].legend(fontsize = 14)
    ax[i].plot(strain_total['Cp_con'], strain_total[strain], marker = 'o', lw = 0, markersize = 4, color = 'grey') ### plot gray points of OD data.
    ax[i].vlines(15,0.01,2, linestyle = '--')
ax[1].set_xlabel('Cp concentration (µg/mL)', fontsize = 18)
ax[0].set_ylabel('OD$_{600}$', fontsize = 18)
plt.show()

# save_path = './FigureS3.pdf'
# plt.savefig(save_path + '/Mix.pdf', transparent = True, bbox_inches = 'tight')
# plt.close()
```

## Figure S7

MIC test of ribosome reporter strain, YK0136 and its resistance-gene-deleted strain, YK0138.

### Extract the strain name

```python
main_path = '.'
branch_path = '/MIC_data/FigureS9'
folder_path = main_path + branch_path

Folder_list=os.listdir(folder_path)
if '.DS_Store' in Folder_list:
    Folder_list.remove('.DS_Store')
Folder_list[0], Folder_list[1] = Folder_list[1], Folder_list[0] ### change the order to ['YK0136', 'YK0138']
```

### import the MIC results and plot figures

```python
min_func = lambda x: max(x, 0.01) ### constructed function for map function
fig, ax = plt.subplots(ncols = 2, nrows = 1,figsize = (8,3), sharey = True)
for i, strain in enumerate(Folder_list):
    data_path = folder_path + '/' + strain
    Day_list = os.listdir(data_path)
    strain_total = pd.DataFrame({'Cp_con':[], strain:[]})
    for day in Day_list:
        day_data = pd.read_csv(data_path + '/' + day)
        strain_total = pd.concat([strain_total, day_data])
    strain_average = strain_total.groupby('Cp_con', as_index = False).mean()
    strain_total[strain] = strain_total[strain].map(min_func)
    ax[i].set_xlim(0.1,1000)
    ax[i].set_ylim(0.01,2)
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')
    ax[i].plot(strain_average['Cp_con'], strain_average[strain], color = 'black', label = strain, lw =2) ### draw black lines of average OD data.
    ax[i].legend(fontsize = 14)
    ax[i].plot(strain_total['Cp_con'], strain_total[strain], marker = 'o', lw = 0, markersize = 4, color = 'grey') ### plot gray points of OD data.
    ax[i].vlines(15,0.01,2, linestyle = '--')
ax[0].set_ylabel('OD$_{600}$', fontsize = 18)

fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.xlabel('Cp concentration (µg/mL)', fontsize = 18)

plt.show()

# save_path = './FigureS9.pdf'
# plt.savefig(save_path + '/Mix.pdf', transparent = True, bbox_inches = 'tight')
# plt.close()
```
